```python
# import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# dataset
data = [11, 12, 13, 14, 15, 16, 17, 18, 19, 19, 590, 18, 17, 587, 16, 15, 14, 13, 12, 521, 11]
```

```python
# identify outliers by using charts/visuals --> histogram, scatter plot, box plot
# 1] histogram
plt.hist(data)
```

```python
# 2] scatter plot
plt.scatter(range(len(data)), data)
```

```python
# 3] boxplot
sns.boxplot(data)
```

# Z-Score Method

```python
outlier = []

def detect_outlier(n):
    threshold = 2
    mean = np.mean(n)
    std = np.std(n)  # Standard_deviation

    for i in n:
        z_score = (i-mean)/std

        if abs(z_score) > threshold:
            outlier.append(i)
    return outlier
```

```python
print(f'Detected Outlier:{detect_outlier(data)}')
```

# IQR

```python
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
print(f'Q1: {q1} \nQ3: {q3}')
```

```python
IQR = q3 - q1
print(f'IQR: {IQR}')
```

```python
lower_fence = q1 - 1.5*IQR
upper_fence = q3 + 1.5*IQR
print(f'lower_fence: {lower_fence} \nupper_fence: {upper_fence}')
```

```python
print('-------------   Clean Data   -------------')
for x in data:
    if x >= lower_fence and x <= upper_fence:
        print(x, end=' ')
```
