1. Extracting Specific Characters
   Scenario: You are given a product code in the format `PROD2024XYZ`.  
   Task: Write a script to extract the following:
   
   The first three characters (product type).
   The year (`2024`).
   Last three characters.

   Input: "PROD2024XYZ"  

```python
Input= "PROD2024XYZ"
print(f'First 3 character: {Input[:3]}')
print(f'Year:{Input[4:8]}')
print(f'Last 3 character: {Input[-3:]}')
```

4. Replacing Characters
   Scenario: Your company uses a unique identifier for employees in the format `EMP12345`. You want to anonymize the identifier by replacing digits with `X`.   
   Task: Write a script to transform "EMP12345" into "EMPXXXXX".  

```python
task = "EMP12345" 
result = task[:3] + 'XXXXX'
print(result)
```

```python
# another way
for i in "12345":
    result = result.replace(i, 'X')
print(result)
```

8. Palindrome Check
   Scenario: Given a string, check if it's a palindrome (reads the same forwards and backwards). Ignore case sensitivity.  
   
   Input: "Radar" 

```python
Input = "Radar"

if Input.lower() == Input.lower()[::-1]:
    print("Yes, Palindrome")
else:
    print("No, Palindrome")
```

10. Extracting the Middle Character(s)
    Scenario: For a given string, extract the middle character if the string length is odd. If even, extract the two middle characters.  
   
    Input: "Python"  

```python
Input = "Python"
ans = len(Input)
if ans%2 == 0:
    print(Input[ans//2 - 1 : ans//2 + 1])
else:
    print(Input[ans//2])
```

12. Masking Sensitive Information
    Scenario: You have a credit card number, and only the last 4 digits should remain visible.  
    Task: Write a script to mask the rest with `*`.  

    Input: "1234567812345678" 

```python
Card = "1234567812345678"
masked = '*' * (len(Card) - 4) + Card[-4:]
print(masked)
```


13. Checking String Symmetry
    Scenario: You need to check if a string has the same characters on the left and right when split in the middle.  
    Task: Write a script to verify symmetry.  

    Input: "abccba" 

```python
Input = "abccba"
mid = len(Input) // 2
if Input[:mid] == Input[mid:][::-1]:
    print('Symmetry')
else:
    print('Not Symmetry')
```

16. Removing Specific Characters
    Scenario: A string contains unwanted characters like `#` or `@`.  
    Task: Write a script to remove all `#` and `@` symbols.  

    Input: "Hello@World#2024" 

```python
Input = "Hello@World#2024"
clean = Input.replace('@', ' ').replace('#', ' ')
print(clean)
```

17. Extracting Capital Letters
    Scenario: A string contains a mix of uppercase and lowercase letters.  
    Task: Extract only the capital letters.  

    Input: "PyThOnIsFuN"

```python
Input = "PyThOnIsFuN"
clean = []
for i in Input:
    if i in i.upper():
        clean.append(i)
print(clean)
```
